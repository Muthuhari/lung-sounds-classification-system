```python
#read Data in Csv file

import pandas as pd
df_path=('E:/EDA/patient_diagnosis.csv')
df=pd.read_csv(df_path)
df.head()
```

```python
#Catogorized Conditions 

df.Condition.values
df.groupby('Condition')['Condition'].agg('count')
```

```python
# importng text files
import os
# save all text files names to files_text variable in the data set and print 
audio_text_path='E:/EDA/audio_and_txt_files'
files_text=[text.split('.')[0] for text in os.listdir(path=audio_text_path) if '.txt' in text]
files_text
```

```python
#extracting text file list to tokens

def token_List(file_name):
    return file_name.split('_')
```

```python
All_Text_file=[]
for f in files_text:
    df1=pd.read_csv(audio_text_path + '/' + f +'.txt', sep='\t',names=['start','end','crackles','wheezes'])
    df1['file_name']=f
    
    #file name features 
    
    file_name_features = token_List(f)
    df1['Patient_number']=file_name_features[0]
    df1['acquisition_mode']=file_name_features[3]
    
    All_Text_file.append(df1)
    
Text_Data_Frame=pd.concat(All_Text_file)
Text_Data_Frame.reset_index()
```
